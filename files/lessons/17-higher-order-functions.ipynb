## ðŸ“– Introduction: Functions Are Values

In Python, a **function is just another kind of value** â€” you can store it in a variable, put it in a list, or pass it to another function!

A **higher-order function** is a function that:
- **Accepts** another function as an argument, OR
- **Returns** a function as its result

The key insight: no parentheses `()` means you're *referring* to the function; parentheses mean you're *calling* it.

```python
def say_hello():
    print("Hello!")

action = say_hello   # storing the function â€” no ()
action()             # NOW we call it
```
